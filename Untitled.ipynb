```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('customer_data.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.set_index('Customer ID',inplace=True)
```

```python
data.index
```

```python
features = data[['Annual Income', 'Spending Score']]
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)
```

```python
# Plot the elbow method
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
optimal_clusters = 5
```

```python
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(scaled_features)
```

```python
cluster_labels
```

```python
sil_score = silhouette_score(scaled_features, cluster_labels)
print(f'Silhouette Score: {sil_score}')
```

```python
data['Cluster'] = cluster_labels
```

```python
data['Cluster']
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Annual Income', y='Spending Score', hue='Cluster', palette='viridis')
plt.title('Customer Segments')
plt.xlabel('Annual Income ($)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

```python
filtered_data = data[(data['Cluster'] == 1) | (data['Cluster'] == 2)]
targetCustomerID = filtered_data.index.tolist()
```
